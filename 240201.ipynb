```python
#to convert training annotations and organize them into YOLOv5 format

from PIL import Image
from shutil import copyfile
import os

a = "/content/drive/MyDrive/project/unzipped/TsignRecgTest1994Annotation.txt"
src = "/content/drive/MyDrive/project/unzipped"
img_o = "/content/drive/MyDrive/project/yolo_dataset/images/val"
lab_o = "/content/drive/MyDrive/project/yolo_dataset/labels/val"

os.makedirs(img_o, exist_ok=True)
os.makedirs(lab_o, exist_ok=True)

with open(a, 'r') as f:
    for l in f:
        p = l.strip().split(';')
        if len(p) < 8: continue
        n = p[0]
        x1, y1, x0, y0 = int(p[1]), int(p[2]), int(p[5]), int(p[6])
        c = int(p[7])
        ip = os.path.join(src, n)
        try:
            with Image.open(ip) as im: w, h = im.size
        except: continue
        xc = ((x0 + x1) / 2) / w
        yc = ((y0 + y1) / 2) / h
        bw = (x1 - x0) / w
        bh = (y1 - y0) / h
        lp = os.path.join(lab_o, n.replace(".png", ".txt"))
        with open(lp, 'w') as lf:
            lf.write(f"{c} {xc:.6f} {yc:.6f} {bw:.6f} {bh:.6f}\n")
        copyfile(ip, os.path.join(img_o, n))
```

```python
import os
from PIL import Image

def crop_imgs(img_dir, lbl_dir, out_dir):
    os.makedirs(out_dir, exist_ok=True)
    for lbl_file in os.listdir(lbl_dir):
        if not lbl_file.endswith('.txt'):
            continue
        img_path = os.path.join(img_dir, lbl_file.replace('.txt', '.png'))
        if not os.path.exists(img_path):
            continue
        img = Image.open(img_path)
        w, h = img.size

        with open(os.path.join(lbl_dir, lbl_file)) as f:
            for i, line in enumerate(f):
                c, x, y, bw, bh = map(float, line.strip().split())
                left = int((x - bw/2) * w)
                top = int((y - bh/2) * h)
                right = int((x + bw/2) * w)
                bottom = int((y + bh/2) * h)
                crop = img.crop((left, top, right, bottom))

                class_dir = os.path.join(out_dir, str(int(c)))
                os.makedirs(class_dir, exist_ok=True)
                out_path = os.path.join(class_dir, f"{lbl_file[:-4]}_{i}.png")
                crop.save(out_path)

crop_imgs(
    img_dir="/content/drive/MyDrive/project/yolo_dataset/images/train",
    lbl_dir="/content/drive/MyDrive/project/yolo_dataset/labels/train",
    out_dir="/content/drive/MyDrive/project/classification_data/train"
)

crop_imgs(
    img_dir="/content/drive/MyDrive/project/yolo_dataset/images/val",
    lbl_dir="/content/drive/MyDrive/project/yolo_dataset/labels/val",
    out_dir="/content/drive/MyDrive/project/classification_data/val"
)
```

```python
valid_exts = ('.jpg', '.jpeg', '.png', '.bmp')

x, y = [], []
for label in os.listdir(data_dir):
    cls_path = os.path.join(data_dir, label)
    for img_name in os.listdir(cls_path):
        if not img_name.lower().endswith(valid_exts):
            continue
        img_path = os.path.join(cls_path, img_name)
        img = cv2.imread(img_path)
        if img is None:
            print(f"Skipped unreadable image: {img_path}")
            continue
        img = cv2.resize(img, (img_size, img_size))
        x.append(img)
        y.append(int(label))
```

```python
import os
import cv2

def load_data(split_dir, img_size=64):
    x, y = [], []
    skipped = 0

    for label in sorted(os.listdir(split_dir)):
        cls_path = os.path.join(split_dir, label)
        if not os.path.isdir(cls_path):
            continue
        for img_name in os.listdir(cls_path):
            img_path = os.path.join(cls_path, img_name)
            img = cv2.imread(img_path)
            if img is None:
                skipped += 1
                continue
            img = cv2.resize(img, (img_size, img_size))
            x.append(img)
            y.append(int(label))

    print(f"Loaded {len(x)} images from {split_dir}")
    print(f"Skipped {skipped} unreadable images")
    return x, y
```

```python
train_dir = '/content/drive/MyDrive/project/classification_data/train'
val_dir = '/content/drive/MyDrive/project/classification_data/val'

x_train, y_train = load_data(train_dir)
x_val, y_val = load_data(val_dir)
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical

# Convert lists to numpy arrays
x_train = np.array(x_train, dtype='float32') / 255.0
x_val = np.array(x_val, dtype='float32') / 255.0
y_train = to_categorical(y_train, num_classes=58)
y_val = to_categorical(y_val, num_classes=58)

print("Shapes:")
print("x_train:", x_train.shape)
print("y_train:", y_train.shape)
print("x_val:", x_val.shape)
print("y_val:", y_val.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(64,64,3)),
    BatchNormalization(),
    MaxPooling2D(),

    Conv2D(64, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(),

    Conv2D(128, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(),

    Flatten(),
    Dropout(0.5),

    Dense(256, activation='relu'),
    Dropout(0.5),

    Dense(58, activation='softmax')  # for 58 classes
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

callbacks = [
    #EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3)
]

from sklearn.utils.class_weight import compute_class_weight
import numpy as np

y_train_labels = np.argmax(y_train, axis=1)

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_labels),
    y=y_train_labels
)

class_weights = dict(enumerate(class_weights))

model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=100, callbacks=callbacks,class_weight=class_weights)
```

```python
val_loss, val_acc = model.evaluate(x_val, y_val, verbose=1)
print(f"Validation Accuracy: {val_acc:.4f}")
print(f"Validation Loss: {val_loss:.4f}")
```

```python
def scale_coords(img1_shape, coords, img0_shape, ratio_pad=None):
    if ratio_pad is None:
        gain = min(img1_shape[0] / img0_shape[0], img1_shape[1] / img0_shape[1])
        pad = ((img1_shape[1] - img0_shape[1] * gain) / 2,
               (img1_shape[0] - img0_shape[0] * gain) / 2)
    else:
        gain = ratio_pad[0][0]
        pad = ratio_pad[1]
    coords[:, [0, 2]] -= pad[0]
    coords[:, [1, 3]] -= pad[1]
    coords[:, :4] /= gain
    coords[:, :4] = coords[:, :4].round()
    return coords
```

```python
import torch
import numpy as np
import cv2
from utils.augmentations import letterbox
from utils.general import non_max_suppression
from models.experimental import attempt_load
from tensorflow.keras.models import load_model

m1 = attempt_load('/content/drive/MyDrive/project/best.pt')
m2 = load_model('/content/drive/MyDrive/project/cnn_model.h5')

def detect_and_classify(p):
    i0 = cv2.imread(p)
    i = letterbox(i0, 640, stride=32, auto=False)[0]
    i = i.transpose(2, 0, 1)[::-1]
    i = np.ascontiguousarray(i)
    t = torch.from_numpy(i).float().unsqueeze(0) / 255.0
    with torch.no_grad():
        p = m1(t)[0]
        d = non_max_suppression(p)[0]
    res = []
    for *xyxy, conf, cls in d:
        b = scale_coords(t.shape[2:], torch.tensor([xyxy]), i0.shape).round()[0].int().tolist()
        x1, y1, x2, y2 = b
        c = i0[y1:y2, x1:x2]
        if c.size == 0: continue
        r = cv2.resize(c, (64,64)) / 255.0
        r = np.expand_dims(r, axis=0)
        p = m2.predict(r)
        res.append((cls.item(), int(np.argmax(p)), b))
    return res, i0
```

```python
import sys
sys.path.insert(0, '/content/yolov5')

import os
import cv2
import torch
import numpy as np
from flask import Flask, render_template, request
from tensorflow.keras.models import load_model
from utils.general import non_max_suppression
from utils.augmentations import letterbox
from models.experimental import attempt_load

app = Flask(__name__, template_folder='/content/templates')
os.makedirs('static/uploads', exist_ok=True)

yolo = attempt_load('/content/drive/MyDrive/project/best.pt')
cnn = load_model('/content/drive/MyDrive/project/cnn_model.h5')

def classify(i):
    i = cv2.resize(i, (64, 64))
    i = np.expand_dims(i, 0).astype('float32') / 255.0
    p = cnn.predict(i)
    return int(np.argmax(p))

@app.route('/', methods=['GET', 'POST'])
def index():
    r = []
    u = None

    if request.method == 'POST':
        f = request.files['image']
        p = 'static/uploads/' + f.filename
        f.save(p)
        u = p

        img0 = cv2.imread(p)
        img = letterbox(img0, 640, stride=32)[0]
        img = img.transpose(2, 0, 1)[::-1]
        img = np.ascontiguousarray(img)
        img = torch.from_numpy(img).float().unsqueeze(0) / 255.0

        with torch.no_grad():
            pred = yolo(img)[0]
            det = non_max_suppression(pred, 0.3, 0.45)[0]

        if det is not None and len(det):
          for *xyxy, conf, cls in det:
              b = [int(x.item()) for x in xyxy]
              x1, y1, x2, y2 = b

              h, w = img0.shape[:2]
              x1 = max(0, min(x1, w - 2))
              x2 = max(0, min(x2, w - 2))
              y1 = max(0, min(y1, h - 2))
              y2 = max(0, min(y2, h - 2))

              cimg = img0[y1:y2, x1:x2]
              if cimg.size == 0: continue
              cl = classify(cimg)
              r.append({'box': [x1, y1, x2, y2], 'yolo_class': int(cls.item()), 'cnn_class': cl})


    return render_template('index.html', result=r, image=u)

if __name__ == '__main__':
    app.run(port=5001)
```
